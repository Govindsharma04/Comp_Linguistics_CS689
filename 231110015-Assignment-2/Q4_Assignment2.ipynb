```python
# there are 5 classes given in question and as given in sir's lecture slides we are dividing them in 2n-1 classes which is 9 here
classes=['B-PER','I-PER','B-LOC','I-LOC','B-ORG','I-ORG','B-MISC','I-MISC','O']

# function to calculate precision recall and f-score for each class saperately
def pre_rec_fsc(ground_truth ,predicted_labels):
    # Initializing the dictionaries to store precision recall and f1-score respectively for each class
    precision_dict = {}
    recall_dict = {}
    f1_dict = {}
    macro_F1_score = 0
    # For better comparison, storing the ground_truth and predicted labels in a single list
    true_labels_flatten = [label for sentence in ground_truth for label in sentence]
    predicted_labels_flatten = [label for sentence in predicted_labels for label in sentence]

    # Iterating over the class one by one
    for class_name in classes:
        # Initializing the variables to count true positives, false positives, and false negatives for each class saperately
        true_positive = 0
        false_positive = 0
        false_negative = 0

        # Iterating over ground_truth labels and predicted labels to calculate TP(true positive), FP(falsw positive), and FN(false negative)
        for true_label, pred_label in zip(true_labels_flatten, predicted_labels_flatten):
            if true_label == class_name and pred_label == class_name:
                true_positive += 1
            elif true_label == class_name and pred_label != class_name:
                false_negative += 1
            elif true_label != class_name and pred_label == class_name:
                false_positive += 1

        # Calculating Precision
        if (true_positive + false_positive) != 0:
          precision = true_positive / (true_positive + false_positive)
        else:
          precision = 0
        # Calculating Recall
        if (true_positive + false_negative) != 0:
          recall = true_positive / (true_positive + false_negative)
        else:
          recall = 0
        # Calculating F1-score
        if (precision + recall) != 0:
          f1_score = (2 * precision * recall) / (precision + recall)
        else:
          f1_score = 0

        # Storing the precision recall and f1-score in their respective dictionaries one by one
        precision_dict[class_name] = precision
        recall_dict[class_name] = recall
        f1_dict[class_name] = f1_score
        macro_F1_score += f1_score
    # Printing the Precision, Recall, and F1-score for each class
    # and also the macro F1-score
    for class_name in classes:
        precision = precision_dict[class_name]
        recall = recall_dict[class_name]
        f1_score = f1_dict[class_name]
        print(f"Class Name:- {class_name}")
        print(f"Precision recall and F1-score for {class_name} are respectively :- [{precision:.4f}, {recall:.4f}, {f1_score:.4f}]")
        print()
    print(f"Macro_F1-score :- {macro_F1_score/9:.4f}")
```
